Question 5 - Create a new column named “num tokens”. Compute the number of tokens (words) in each sample and store the results in the newly created column. Reassign the modified dataset to the same variable “ds”. How many tokens (in millions) are there in the entire dataset?

```python
ds = load_from_disk("pyarrow_dataset/naamapadamta")
def count_tokens(example):
    return {"num_tokens": len(example["tokens"])}
ds = ds.map(count_tokens)
total_tokens = sum(ds["train"]["num_tokens"]) + sum(ds["test"]["num_tokens"]) + sum(ds["validation"]["num_tokens"])
print(total_tokens)
total_tokens_in_millions = total_tokens / 1_000_000
print(f"Total tokens in the entire dataset (in millions): {total_tokens_in_millions}")
```

Question 6 - The statement that the modified dataset increased the disk space requirement about two fold is

```python
import os
def get_disk_size(path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            total_size += os.path.getsize(file_path)
    return total_size / (1024 * 1024)
```

```python
print(get_disk_size("pyarrow_dataset/naamapadamta"))
def count_tokens(example):
    return {"num_tokens": len(example["tokens"])}
ds = load_from_disk("pyarrow_dataset/naamapadamta")
ds = ds.map(count_tokens)
ds.save_to_disk("pyarrow_dataset/naamapadamta_modified")
print(get_disk_size("pyarrow_dataset/naamapadamta_modified"))
```

Question 7 -
Concatenate all the samples across the splits in the following order: [train:test:validation]. Currently, each sample contains a list of words (tokens). Create a sentence by joining the individual words (tokens) in each sample using a single white space as a delimiter and store the resulting sample in a new column named “text”. Create a new dataset by removing the columns “ner tags” and “tokens”. Store the new dataset in the same variable “ds”. Enter the total number of samples in “ds”

```python
from datasets import load_from_disk
ds = load_from_disk("pyarrow_dataset/naamapadamta")
ds = datasets.concatenate_datasets([ds["train"], ds["test"], ds["validation"]], axis=0)
def create_text_column(example):
    example["text"] = " ".join(example["tokens"])
    return example
ds = ds.map(create_text_column)
ds = ds.remove_columns(["ner_tags", "tokens"])
total_samples = len(ds)
print(f"Total number of samples in the modified dataset: {total_samples}")
```

Question 10 -
Download all the splits of Tamil sub-dataset “inltkh.ta” of https://huggingface.co/datasets/ai4bharat/indic_glue .

Filter the dataset so that each sample contains at least six words (separated by a single white space).Then Interleave the resultant dataset with the above filtered dataset of naamapadam. Take 80% of samples from naamapadam and 20% from indic glue. Enter the number of samples after interleaving the datasets.
Note: Set the value of the seed argument to 42.

```python
ds = load_dataset("ai4bharat/naamapadam", name="ta", split="train+test+validation")
def add_num_tokens(example):
    example["num_tokens"] = len(example["tokens"])
    return example
ds = ds.map(add_num_tokens)
ds = ds.filter(lambda x: x["num_tokens"] >= 6)
ds
```

```python
new_ds = load_dataset("ai4bharat/indic_glue", name="inltkh.ta", split="train+test+validation")
new_ds
```

```python
new_ds = new_ds.filter(lambda x: len(x["text"].split(" ")) >= 6)
new_ds
```

```python
from datasets import interleave_datasets
interleave_eg = interleave_datasets([ds, new_ds], probabilities=[0.8, 0.2], seed=42)
print(interleave_eg)
```
