# L5 - Common Methods

```python
print("Before Filtering")
print(20*"-")
print(imdb)
num=100
imdb_filter100 = imdb.filter(lambda eg:len(eg['text'].split(" "))>=num)
print("After Filtering")
print(20*"-")
print(imdb_filter100)
```

```python
def add_prefix(example):
  example['text'] = "IMDB: " + example['text']
  return example

imdb_prefix = imdb.map(add_prefix)
# print(imdb_prefix)
print(imdb_prefix['train']['text'][100])
```

```python
rotten = load_dataset("cornell-movie-review-data/rotten_tomatoes", split="all")
print(rotten)
print(rotten.features)
```

```python
concat_dataset = datasets.concatenate_datasets([imdb_train, rotten], axis=0)
print(concat_dataset)
```

```python
from datasets import interleave_datasets
interleave_eg = interleave_datasets([imdb_train, rotten], probabilities=[0.6, 0.4])
print(interleave_eg)
```

```python
imdb_iterable = load_dataset("stanfordnlp/imdb", split="train", streaming=True)
print(imdb_iterable)
```
